```python
import cv2
import os
import glob
import numpy as np
import pandas as pd
from imutils import paths
from skimage import img_as_ubyte  #Convert an image to unsigned byte format, with values in [0, 255].
from skimage.filters import gaussian, sobel
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
#converting images to gray and adding gaussian blur
```

```python
def smoothing(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    smoothimg = img_as_ubyte(cv2.GaussianBlur(gray, (5, 5), 0,borderType=cv2.BORDER_CONSTANT))
    return smoothimg
```

```python
#applying canny edge detection
```

```python
def canny_edge_detection(smoothimg, threshold1, threshold2):
    canny_edges = cv2.Canny(smoothimg, threshold1, threshold2)
    return canny_edges
```

```python
#applying sobel edge detection
```

```python
def sobel_edge_detection(smoothimg):
    x = cv2.Sobel(smoothimg, cv2.CV_64F, 1,0, ksize=3, scale=1)
    y = cv2.Sobel(smoothimg, cv2.CV_64F, 0,1, ksize=3, scale=1)
    absx = cv2.convertScaleAbs(x)
    absy = cv2.convertScaleAbs(y)
    sobel_edges = cv2.addWeighted(absx, 0.5, absy, 0.5,0)
    return sobel_edges
```

```python
#canny_sobel_fusion
def edge_fusion(canny_edges, sobel_edges):
    # Perform element-wise maximum to fuse the edge images
    fused_edges = np.maximum(canny_edges, sobel_edges)
    return fused_edges
```

```python
#inserting file path
originpath = r"E:\NSLD\NumericalSign\Train\images1"
print(os.listdir(originpath))
```

```python
#changing current directory
os.chdir(r'E:\NSLD\NumericalSign\Train\images1')
```

```python
#output directory
rsfolder = r"E:\NSLD\NumericalSign\Train\images"
```

```python


#inserting image through loop
for image in os.listdir(originpath):
    path = os.path.join(originpath,image)
    img= cv2.imread(path,1)
    
    #calculation threshold for canny
    sigma = 0.3
    median = np.median(img)
    threshold1 = int(max(0,(1.0-sigma)*median)) 
    threshold2 = int(min(0,(1.0-sigma)*median))
    
    smoothimg = smoothing(img)
    canny_edges = canny_edge_detection(smoothimg, threshold1, threshold2)
    sobel_edges = sobel_edge_detection(smoothimg)
    fused_edges = img_as_ubyte(edge_fusion(canny_edges, sobel_edges))
    
    output_path = os.path.join(rsfolder,image)
    cv2.imwrite(output_path, fused_edges)
    
   
    
    

   


```
